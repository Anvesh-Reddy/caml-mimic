```python
def preprocess(text):
    ###code to deal with possible nested <code> tags...
    starts = [m.start() for m in re.finditer('<code>', text, flags=re.IGNORECASE)]
    #wow
    ends = [m.end() for m in re.finditer('</code>', text, flags=re.IGNORECASE)]
    if len(starts) > len(ends):
        print(text)
    overlaps_w_next = [starts[i+1] < ends[i] for i in range(len(starts)-1)]
    overlaps_w_next.append(False)
    i = 0
    get_new_start = True
    itvs = []
    while i < len(starts):
        itv = starts[i]
        if get_new_start:
            start = itv
        overlaps = overlaps_w_next[i]
        if not overlaps:
            if i >= len(ends):
                itvs.append((start, len(text)))
                break
            else:
                itvs.append((start, ends[i]))
        get_new_start = not overlaps
        i += 1
    for itv in itvs[::-1]:
        text = text[:itv[0]] + text[itv[1]:]
    text = text.lower()
    text = strip_tags(text)
    text = text.replace('\n', ' ')
    text = ' '.join(word_tokenize(text))
    text = '"' + text + '"'
    return text
```

# Concatenate notes with labels

```python
def next_labels(labels_file):
    with open(labels_file, 'r') as lf:
        lr = csv.reader(lf)
        #header
        next(lr)
        fl = next(lr)
        cur_id = int(fl[0])
        cur_labels = [fl[1]]
        
        for row in lr:
            q_id = int(row[0])
            tag = row[1]
            if q_id != cur_id:
                yield cur_id, cur_labels
                cur_labels = [tag]
                cur_id = q_id
            else:
                cur_labels.append(tag)
        yield cur_id, cur_labels
```

```python
with codecs.open('%s/Questions_proc.csv' % SO_DIR, 'r', 'utf-8') as f:
    with codecs.open('%s/questions_labeled.csv' % SO_DIR, 'w', 'utf-8') as of:
        labels_gen = next_labels('%s/Tags.csv' % SO_DIR)
        #header
        next(f)
        of.write(','.join(['ID', 'TEXT', 'LABELS']) + '\n')
        for i,row in tqdm(enumerate(f)):
            row = row.rstrip().split(',')
            try:
                q_id = int(row[0])
            except:
                print(i,row)
            text = ','.join(row[1:])
            cur_id, cur_labels = next(labels_gen)
            if cur_id == q_id:
                of.write(','.join([str(cur_id), text, ';'.join(cur_labels)]) + '\n')
```

# Filter to top 4k codes

```python
with open('%s/questions_labeled.csv' % SO_DIR, 'r') as f:
    with open('%s/questions_labeled_4k.csv' % SO_DIR, 'w') as of:
        r = csv.reader(f)
        w = csv.writer(of)
        #header
        w.writerow(next(r))
        i = 0
        for row in tqdm(r):
            length = len(row[1].split())
            inst_codes = set(str(row[2]).split(';'))
            filtered_codes = inst_codes.intersection(codes)
            if len(filtered_codes) > 0 and length > 100:
                w.writerow(row[:2] + [';'.join(filtered_codes), length])
                i += 1
```

# Split data

```python
tr_name = '%s/train_4k_split.csv' % SO_DIR
dv_name = '%s/dev_4k_split.csv' % SO_DIR
te_name = '%s/test_4k_split.csv' % SO_DIR
```

```python
split = [0.8, 0.1, 0.1]
```

```python
def flip_coin(split):
    val = random.random()
    if val < split[0]:
        return "train"
    elif val >= split[0] and val < sum(split[:2]):
        return "dev"
    else:
        return "test"
```

```python
trf = open(tr_name, 'w')
dvf = open(dv_name, 'w')
tef = open(te_name, 'w')
trw = csv.writer(trf)
dvw = csv.writer(dvf)
tew = csv.writer(tef)
trw.writerow(['ID', 'TEXT', 'LABELS', 'length'])
dvw.writerow(['ID', 'TEXT', 'LABELS', 'length'])
tew.writerow(['ID', 'TEXT', 'LABELS', 'length'])
tri = 0
with open('%s/questions_labeled_4k.csv' % SO_DIR, 'r') as f:
    r = csv.reader(f)
    next(r)
    for row in tqdm(r):
        which = flip_coin(split)
        if which == 'train':
            if tri < 365192:
                trw.writerow(row)
                tri += 1
        elif which == 'dev':
            dvw.writerow(row)
        elif which == 'test':
            tew.writerow(row)
trf.close()
dvf.close()
tef.close()
```
