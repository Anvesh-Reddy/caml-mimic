```python
import json
import os

%pylab --no-import-all inline
import matplotlib.pyplot as plt
%pylab inline
```

```python
basedir = "models/"
```

```python
def plot_metric(iterations, metrics, names):
    """
        up to 4 names only
    """
    fig,ax = plt.subplots(2,2)
    fig.set_size_inches(14,14)
    ax[0][0].set_xlabel("iterations")
    ax[0][1].set_xlabel("iterations")
    ax[1][0].set_xlabel("iterations")
    ax[1][1].set_xlabel("iterations")
    
    if len(names) > 4:
        print("booooo")
        return
    for i,name in enumerate(names):
        p_x = i / 2
        p_y = i % 2
        ax[p_x][p_y].plot(iterations, metrics[name])
        ax[p_x][p_y].plot(iterations, metrics[name], 'b*')
#         ax[p_x][p_y].xticks(iterations)
        ax[p_x][p_y].set_xlabel("iterations")
        ax[p_x][p_y].set_ylabel(name)
        ax[p_x][p_y].set_title(name)
    plt.show()
```

```python
def plot_metrics(model):
    with open(basedir + model + "/metrics.json", 'r') as metrics_file:
        with open(basedir + model + "/params.json", 'r') as params_file:
            metrics = json.load(metrics_file)
            params = json.load(params_file)
            iterations = [i for i in range(1,6)]
            print("Conv window size: " + params['Conv window size'])
            print("Embedding size: " + params["Embedding size"])
            print("Conv output size: " + params["Conv output size"])
            plot_metric(iterations, metrics, ["acc", "prec", "rec", "f1"])
```

```python
models = ["cnn_Feb_24_04:14","cnn_Feb_24_06:57"]
for model in models:
    plot_metrics(model)
```
